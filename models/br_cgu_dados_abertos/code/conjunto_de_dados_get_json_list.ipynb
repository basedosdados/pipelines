```python
import pandas as pd
import requests
import json
import glob
```

```python
session = requests.Session()
organizacoes_df = pd.read_parquet("./parquet/organizacao.parquet", columns=["id"])
```

```python
TAMANHO_PAGINA = 100
QTD_ORGANIZACOES = len(organizacoes_df)

for idx, row in organizacoes_df.iterrows():
    temp_id = row["id"]
    if glob.glob(f'./json/conjunto/{temp_id}*.json'):
        print("Consulta já realizada anteriormente")
        continue
    
    qtd_conjuntos = int(session.get(f"https://dados.gov.br/api/publico/conjuntos-dados/buscar?tamanhoPagina=0&ownerOrg={temp_id}").json().get("totalRegistros"))
    NUMERO_ITERACOES = qtd_conjuntos // TAMANHO_PAGINA + (0 if qtd_conjuntos % TAMANHO_PAGINA == 0 else 1)
    
    print(f"[{idx + 1}/{QTD_ORGANIZACOES}]")
    print(f"Consultando a organização com id {temp_id}")
    print(f"Total de conjuntos: {qtd_conjuntos}")
    print("-"*10)
    for i in range(0, NUMERO_ITERACOES):
        print(f"Consultando conjuntos {i * TAMANHO_PAGINA} a {((i + 1) * TAMANHO_PAGINA) - 1}")
        temp_query_url = f"https://dados.gov.br/api/publico/conjuntos-dados/buscar?offset={i * TAMANHO_PAGINA}&tamanhoPagina={TAMANHO_PAGINA}&ownerOrg={temp_id}"
        response_json = session.get(temp_query_url).json()
        temp_file_name = f"./json/conjunto/{temp_id}_{i + 1}.json"
        print(f"Exportando os dados para o caminho {temp_file_name}")
        with open(temp_file_name, "w") as f:
            json.dump(response_json, f, indent=4)
        print("-"*10)
    print()
```
