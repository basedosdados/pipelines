```python
import pandas as pd
import json
from glob import glob
from collections import OrderedDict
```

```python
data_list = []
for temp_json_path in glob('./json/organizacao/*.json'):
    with open(temp_json_path, 'r') as f:
        data = json.load(f)
        data_list.extend(data)
```

```python
df = pd.DataFrame.from_records(data_list)
df.drop(columns=["urlImagem"], inplace=True)
df = df[["id", "titulo", "nome", "descricao", "qtdConjuntoDeDados", "qtdSeguidores", "organizationEsfera", "organizationUf", "organizationMunicipio"]]
df
```

```python
df.to_parquet("./parquet/organizacao.parquet", index=False)
```
