```python
import pandas as pd
import json
from glob import glob
from collections import OrderedDict

data_list = []
for temp_json_path in glob('./json/recurso/*.json'):
    with open(temp_json_path, 'r') as f:
        data = json.load(f)
        temp_dict = OrderedDict()
        
        for recurso in data["resources"]:
            temp_dict["id"] = recurso["id"]
            temp_dict["nome"] = recurso["name"]
            temp_dict["descricao"] = recurso["description"]
            temp_dict["id_conjunto"] = recurso["package_id"]
            temp_dict["tipo"] = recurso["tipo"]
            temp_dict["formato"] = recurso["format"]
            temp_dict["tamanho_bytes"] = recurso["size"]
            temp_dict["url_download"] = recurso["url"]
            temp_dict["data_criacao"] = recurso["created"]
            temp_dict["data_modificacao_metadados"] = recurso["metadata_modified"]
            temp_dict["quantidade_downloads"] = recurso["quantidadeDownloads"]
            data_list.append(dict(temp_dict))
            temp_dict.clear()
```

```python
final_df = pd.DataFrame(data_list)
final_df.to_parquet('./parquet/recurso.parquet', index=False)
```
