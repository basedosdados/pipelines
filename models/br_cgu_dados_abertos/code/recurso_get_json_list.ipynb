```python
import pandas as pd
import requests
import json
import glob
```

```python
session = requests.Session()
conjuntos_df = pd.read_parquet("./parquet/conjunto.parquet", columns=["id", "nome"])
```

```python
QTD_CONJUNTOS = len(conjuntos_df)

for idx, row in conjuntos_df.iterrows():
    temp_id = row["id"]
    temp_nome = row["nome"]
    
    if glob.glob(f'./json/recurso/{temp_id}.json'):
        print("Consulta já realizada anteriormente")
        continue
    
    print(f"[{idx + 1}/{QTD_CONJUNTOS}]")
    print(f"Consultando recursos do conjunto {temp_id}")
    temp_query_url = f"https://dados.gov.br/api/publico/conjuntos-dados/{temp_nome}"
    
    try:
        response_json = session.get(temp_query_url).json()
        if "erro" in response_json.keys():
            raise Exception('Erro interno.')
    except:
        print('JSON inválido. Ignorando esta requisição.')
        error_line = f"nome: {temp_nome}, id: {temp_id}\n"
        with open('./recurso_erros.txt', 'r') as f:
            existing_lines = f.readlines()
        if error_line not in existing_lines:
            with open('./recurso_erros.txt', 'a') as f:
                f.write(error_line)
        continue
        
    temp_file_name = f"./json/recurso/{temp_id}.json"
    print(f"Exportando os dados para o caminho {temp_file_name}")
    with open(temp_file_name, "w") as f:
        json.dump(response_json, f, indent=4)
    print()
```
