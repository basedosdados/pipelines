```python
# Lista de colunas fixas (que não vão para long)
id_vars = [
    "sigla_uf",
    "id_municipio",
    "rede",
    "meta_alfabetizacao_2024",
    "meta_alfabetizacao_2025",
    "meta_alfabetizacao_2026",
    "meta_alfabetizacao_2027",
    "meta_alfabetizacao_2028",
    "meta_alfabetizacao_2029",
    "meta_alfabetizacao_2030",
]


# Função auxiliar para "derreter" um grupo de colunas por variável
def melt_por_variavel(df, variavel):
    colunas = [f"{variavel}_2023", f"{variavel}_2024"]
    df_melted = df_2024[id_vars + colunas].melt(
        id_vars=id_vars,
        value_vars=colunas,
        var_name="variavel_ano",
        value_name=variavel,
    )
    # Extrai o ano da string da coluna e transforma em inteiro
    df_melted["ano"] = (
        df_melted["variavel_ano"].str.extract(r"(\d{4})").astype(int)
    )
    df_melted = df_melted.drop(columns="variavel_ano")
    return df_melted


# Faz o melt para cada variável que será empilhada
df_taxa = melt_por_variavel(df_2024, "taxa_alfabetizacao")
df_nivel = melt_por_variavel(df_2024, "nivel_alfabetizacao")
df_participacao = melt_por_variavel(df_2024, "percentual_participacao")

# Junta tudo pelo conjunto de chaves
df_long = df_taxa.merge(df_nivel, on=id_vars + ["ano"]).merge(
    df_participacao, on=id_vars + ["ano"]
)
```
